```python
import numpy as np
import matplotlib.pyplot as plt

# XOR Dataset

# generate random numbers between -1, 1 in both X and Y
random_numbers = np.array(np.random.uniform(-1, 1, (1000, 2)))
plt.scatter(random_numbers[:,0], random_numbers[:,1])
plt.title('Random numbers without classification')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
# classify the points:
# x1 > 0 and x2 > 0 : 0
# x1 < 0  and x2 < 0 : 0

# else : 1

colors = ['red' if (x1 > 0 and x2 > 0) or (x1 < 0 and x2 <0) else 'green' for x1, x2 in random_numbers]

# make a scatterplot with color coding
plt.scatter(random_numbers[:,0], random_numbers[:,1], c=colors)
plt.title('Random numbers with classificaiton')
plt.xlabel('x1')
plt.ylabel('x2')
```

```python
# Donut dataset

# create random numbers x1,x2 

# create a dataframe from them

# add x1^2, x2^2, and x1*x2

# save to CSV without header, row, column


from sklearn.datasets import make_circles
import pandas as pd

circles, labels = make_circles(1000)
colors = ['red' if labels[index] == 1 else 'green' for index, element in enumerate(circles)]
plt.scatter(circles[:,0], circles[:,1], c=colors)
df = pd.DataFrame(circles, columns=['x', 'y'])
df['label'] = labels
df.head()
df.to_csv('donut_data.csv', header=False, index=False)
```

```python
# simple edge detection using convolution
from scipy.signal import convolve2d
from PIL import Image
image = Image.open('lena.jpeg')
# convert to grayscale
gray = np.mean(image, axis=2)
# matrixes for edgedetection
H1 = np.array([[1, 0, -1], [2, 0, -2], [1, 0 , -1]])
H2 = np.array([[1, 2, 1], [0, 0, 0], [-1, -2 , -1]])
# convolve the matrix with the gray-scale image

Gx = convolve2d(gray, H1)
Gy = convolve2d(gray, H2)
# take the square root of the sum of the convolved arrays squared
G = np.sqrt(np.multiply(Gx,Gx) + np.multiply(Gy,Gy))
plt.imshow(G,cmap='gray')
```
